## Examples

````python

# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_blobs

# --- Supervised Learning (Synthetic Classification Data) ---
# X, y = make_classification(n_samples=100, n_features=2, n_classes=2, random_state=42)
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title("Supervised Learning: Classification")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# --- Unsupervised Learning (Synthetic Clustering Data) ---
X_blobs, _ = make_blobs(n_samples=100, centers=3, n_features=2, random_state=42)
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c='gray')
plt.title("Unsupervised Learning: Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Reinforcement Learning is usually not shown with static datasets
print("Reinforcement Learning is agent-based and requires environments like OpenAI Gym.")
````

## Lets explore the example a little bit 

This generates a synthetic dataset for classification tasks

```python 
X, y = make_classification(
    n_samples=100,         # Number of samples (rows)
    n_features=2,          # Number of features (columns)
    n_informative=2,       # Number of informative features (actually useful for classification)
    n_redundant=0,         # Number of redundant features (linear combinations of informative ones)
    n_clusters_per_class=1,# Number of clusters per class
    random_state=42        # Seed for reproducibility
)
```
### What is `X`?
It is a 2D array with 100 samples and 2 features 

### What is `y`?
It is a 1D array with 100 labels (0 or 1), representing the class for each sample. 

This is useful for visualizing and testing classification algorithms with simple controlled data. 


````python
print("Type of X:", type(X))
print("Shape of X:", X.shape)
print("Type of y:", type(y))
print("Shape of y:", y.shape)
````
